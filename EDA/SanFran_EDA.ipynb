```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
pd.set_option('display.max_rows', None)

df = pd.read_parquet("../../data/cleanedSanFranData.parquet")
```

```python
# reindex dataframe
df.set_index("requestId", inplace=True)
print(df.info())
df.head()
```

```python
# replace NaNs in category columns to Missing
df2 = df.copy()
df2["status"] = df["status"].cat.add_categories("Missing")
df2["source"] = df["source"].cat.add_categories("Missing")
df2[["status", "source"]] = df2[["status", "source"]].fillna("Missing")

print("------- Count of Values in dataSubtype -------")
print(df2["dataSubtype"].value_counts())

print("\n------- Count of Values in Category -------")
print(df2["category"].value_counts())

print("\n------- Count of Values in Status")
print(df2["status"].value_counts())

print("\n------- Count of Values in Source")
print(df2["source"].value_counts())
```

```python
# extract month and year into separate columns
df2.insert(2, "Month", df2.dateTime.dt.month)
df2.insert(3, "Day", df2.dateTime.dt.day_name())
df2.insert(4, "Year", df2.dateTime.dt.year)
df2.insert(5, "Hour", df2.dateTime.dt.hour) # I'm not sure if this is in UTC
```

```python
df2.head(2)
```

```python
categories = df2['category'].value_counts()[:20]
categories.name = 'count'
ax = categories.plot(kind="barh", figsize=(16,12), title="Most Frequent Categories in the Dataset")
ax.bar_label(ax.containers[0])
plt.show()
```

### Table View with Percentages of most Frequent Categories

```python
percentages = (df2['category'].value_counts(normalize=True) * 100)[:20]
percentages.name = 'percentage'
table = pd.concat([categories,percentages],axis=1)
table
```
